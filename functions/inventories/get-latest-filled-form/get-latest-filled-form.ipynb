```python
# Get the latest finalized (is_empty=False) inventories document
from pymongo import MongoClient, DESCENDING
import os
from dotenv import load_dotenv
from bson.json_util import dumps  # preserves ObjectId/dates

load_dotenv()

uri = os.getenv("MONGODB_URI")
database = os.getenv("MONGODB_DATABASE")

client = MongoClient(uri, serverSelectionTimeoutMS=5000)
db = client[database]
coll = db["inventories"]

query = {"is_empty": False}
projection = {"_id": 0}

doc = coll.find_one(
    query,
    projection=projection,
    sort=[("created", DESCENDING), ("date", DESCENDING), ("current_form_qr_code", DESCENDING)],
)

if not doc:
    raise RuntimeError("No finalized (is_empty=False) inventory document found.")

print(dumps(doc, indent=2, ensure_ascii=False))
```
