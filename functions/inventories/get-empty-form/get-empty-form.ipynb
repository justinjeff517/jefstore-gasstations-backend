```python
# Get the single inventories document where is_empty=True
from pymongo import MongoClient
import os
from dotenv import load_dotenv
from bson.json_util import dumps  # handles ObjectId, dates, etc.

load_dotenv()

uri = os.getenv("MONGODB_URI")
database = os.getenv("MONGODB_DATABASE")

client = MongoClient(uri)
db = client[database]
coll = db["inventories"]

query = {"is_empty": True}

count = coll.count_documents(query)
if count != 1:
    raise RuntimeError(f"Expected exactly 1 is_empty=True document, found {count}.")

doc = coll.find_one(query, {"_id": 0})  # exclude Mongo's _id for cleaner output
print(dumps(doc, indent=2, ensure_ascii=False))
```
