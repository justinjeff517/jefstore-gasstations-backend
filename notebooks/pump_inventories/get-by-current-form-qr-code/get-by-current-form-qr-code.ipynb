```python
import os, json
from dotenv import load_dotenv
from pymongo import MongoClient
from bson import json_util

load_dotenv()
uri = os.getenv("MONGODB_URI")
database = os.getenv("MONGODB_DATABASE")

client = MongoClient(uri, tz_aware=True)
coll = client[database]["pump_inventories"]

def fetch_pump_inventories_by_qr(code: str):
    q = {"current_form_qr_code": code}
    return list(coll.find(q).sort([("location",1),("dispenser_name",1),("pump_id",1),("date",-1)]))

current_code = "11458998"
docs = fetch_pump_inventories_by_qr(current_code)
print(len(docs))
print(json.dumps(docs, default=json_util.default, ensure_ascii=False, indent=2))
```
