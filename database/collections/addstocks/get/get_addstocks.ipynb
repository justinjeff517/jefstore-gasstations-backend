```python
import os, json
from pymongo import MongoClient
from bson import json_util

try:
    from dotenv import load_dotenv
    load_dotenv()
except:
    pass

uri = os.getenv("MONGODB_URI", "mongodb://localhost:27017")
database = os.getenv("MONGODB_DATABASE", "test")

client = MongoClient(uri)
coll = client[database]["addstocks"]

docs = list(coll.find().sort("date", 1))
print("Total addstocks:", len(docs))

for d in docs:
    print(json.dumps(d, default=json_util.default, ensure_ascii=False, indent=2))
```
