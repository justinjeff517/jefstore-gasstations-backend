```python
import json
from pymongo import MongoClient
import os
from dotenv import load_dotenv
from bson.json_util import dumps  # <-- handles datetime, ObjectId, Decimal128, etc.

load_dotenv()

uri = os.getenv("MONGODB_URI")
database = os.getenv("MONGODB_DATABASE")
collection = "inventories"

client = MongoClient(uri)
db = client[database]
coll = db[collection]

inventories = list(coll.find({}, {"_id": 0}))

# Pretty print JSON (Mongo-safe)
print(dumps(inventories, indent=2))
```
